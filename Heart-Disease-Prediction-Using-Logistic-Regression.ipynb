```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the data set
heart_data = pd.read_csv('heart_disease_data.csv')
heart_data.head()
```

```python
#print last 5 rows
heart_data.tail()
```

```python
#getting info of data 
heart_data.shape
```

```python
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
#analyzing the data
heart_data.describe()
```

```python
#checking the distribution of target variable
heart_data['target'].value_countsounts()
```

1 --> Defective Heart
0 --> Healthy Heart

## Splitting the data

```python
X = heart_data.drop(columns='target',axis=1)
Y = heart_data['target']
```

```python
print(X)
print(Y)
```

## Training and Testing Data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training

```python
model = LogisticRegression()
```

```python
#training The Logistic Regression Model
model.fit(X_train,Y_train)
```

## Model Evaluation

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy On Training Data: {:.2%}".format(training_data_accuracy))
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy On Training Data: {:.2%}".format(test_data_accuracy))
```

## Building The Predictive System

```python
input_data = (54,1,2,120,258,0,0,147,0,0.4,1,0,3)

# change the input data into np array
input_data_nparray = np.asarray(input_data)

# reshape the np array 
input_data_reshaped = input_data_nparray.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] ==  0):
    print("The Person not have any Heart Disease")
else:
    print("The Person have Heart Diesease")
```
